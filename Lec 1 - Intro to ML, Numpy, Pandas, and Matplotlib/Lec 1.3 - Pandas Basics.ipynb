# Series & DataFrames - two basic datastructures in Pandas

series-and-dataframe.width-1200.png


**Series**  
Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.).  
The axis labels are collectively referred to as the index.  
**Pandas Series is nothing but a column in an excel sheet.**

```python
s = pd.Series([1, 2, 3, 4, 5])
s
```

```python
s.index
```

```python
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
s
```

```python
s.index
```

**Series can be instantiated from dicts:**

```python
data = {"a": 0.0, "b": 1.0, "c": 2.0}
pd.Series(data)
```

**From scalar value**

If data is a scalar value, an index must be provided. The value will be repeated to match the length of index.

```python
pd.Series(5.0, index=["a", "b", "c", "d", "e"])
```

**Indexing and Slicing Series**  

```python
s
```

Series acts very similarly to a ndarray and is a valid argument to most NumPy functions. However, operations such as slicing will also slice the index.

```python
s[0], s[1], s[4]
```

```python
s[1:4]
```

A Series is also like a fixed-size dict in that you can get and set values by index label:

```python
s
```

```python
s["a"]
```

```python
s["e"] = 12.0
s
```

**DataFrame**  
DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object.  
**Pandas DataFrame is nothing but an excel sheet.**

**Constructing a DataFrame from a list**

```python
data = [
    [1, 2, 3, 4, 5, 6], 
    ['a', 'b', 'c', 'd', 'e', 'f']
]
pd.DataFrame(data)
```

You can pass the column names

```python
pd.DataFrame(data, columns=['col_1', 'col_2', 'col_3', 'col_4', 'col_5', 'col_6'])
```

**Constructing a DataFrame from a dictionary**

```python
data = {
    'Name' : ['Amal', 'Aneetta', 'Lizzy', 'Jilu'],
    'Job' : ['Engineer', 'Student', 'Teacher', 'Gov Employee']
}

pd.DataFrame(data)
```

**Selecting multiple columns**

```python
df[['Name', 'Age']]
```

**Indexing**  
If we need to select a range of rows, we can specify the range using ":" 


```python
df[0:3]
```

**Selection by label**  
**df.loc** - If we need to select a range of rows, using their labels we can use method loc


```python
df.loc[0:3]
```

Note the difference in output between df.loc[0:3] and df[0:3].  
In **df[0:3]**, the index position between 0 and 3 is selected (3 excluded, array starts at 0).  
But in **df.loc[0:3]**, the lables 0, 1, 2, and 3 are selected.  

**Selection by position**  
If we need to select a range of rows and/or columns, using their positions we can use method iloc:

```python
df.iloc[3]
```

```python
df.iloc[0:3]
```

```python
df.iloc[0:3, 2:5]  # row_start:row_end, col_start:col_end
```

**Boolean indexing**  
Can be used to apply filters to the dataset.  

```python
df[ df.Age> 50] # observe the index, missing indexs indicate that those rows are discarded by the filter.
```

**Handling Missing Data**

```python
data = {
    'Name': ['Amal', 'Aneetta', 'Lizzy'],
    'Age' : [26, np.nan, 58]
}
test_df = pd.DataFrame(data)
```

Check if Null values are present

```python
test_df.info()
```

```python
test_df.isna().sum()
```

**DataFrame.dropna() drops any rows that have missing data:**  
how = {‘any’, ‘all’}, default ‘any’  
Determine if row or column is removed from DataFrame, when we have at least one NA or all NA.  
*‘any’* : If any NA values are present, drop that row or column.  
*‘all’* : If all values are NA, drop that row or column.

```python
test_df.dropna(how="any")
```

**DataFrame.fillna() fills missing data:**

```python
test_df.fillna(value=1000)
```

A better way is to replace the missing values with the mean or median of that column.  
**mean()**, **median()**

```python
mean_age = test_df.Age.mean()
test_df.fillna(value=mean_age)
```

# Data Frames groupby method
Using "group by" method we can:
Split the data into groups based on some criteria
Calculate statistics (or apply a function) to each group


```python
df.groupby(['Sex'])
```

**Calculate mean value for each numeric column per each group**

```python
df.groupby(['Sex']).mean(numeric_only=True)
```

**Once groupby object is created we can calculate various statistics for each group**  
*Note: If single brackets are used to specify the column (e.g. salary), then the output is Pandas Series object. When double brackets are used the output is a Data Frame*


```python
df.groupby('Sex')[['Fare']].mean()
```
